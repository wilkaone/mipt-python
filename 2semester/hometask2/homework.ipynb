```python
def show_images_from_dir(directory, num_images=3):
    image_files = sorted([f for f in os.listdir(directory) if f.endswith('.png')])[:num_images]

    plt.figure(figsize=(12, 4))
    for idx, image_file in enumerate(image_files):
        img = Image.open(os.path.join(directory, image_file))
        plt.subplot(1, num_images, idx + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Image {idx + 1}")
    
    plt.show()
```

```python
show_images_from_dir(save_path, num_images=3)
```
